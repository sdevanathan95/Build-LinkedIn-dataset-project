# Web Scraping

```python
#from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import csv
import io


# getting the code to write the data into the file ready

csv_file = open('Imdb_movie_database2.csv','w', encoding = 'utf-8')
csv_writer = csv.writer(csv_file,delimiter=',')
csv_writer.writerow(['Title','Year','Genre','Description','Starring','Certificate','Runtime','Rating','Votes'])

# Main code for scraping

def webpage(url):
    
    html = requests.get('https://www.imdb.com/'+url).text    # Getting the URL for each page
    i = 0
    Total = 50
    while(html):    
        soup = BeautifulSoup(html)
        container = soup.find('div',class_ = 'lister-list')
        
        # Each try except block pulls one type of information

        for box in container.findAll('div', class_ = 'lister-item mode-advanced'):
            
            try:
                title = box.find('h3',class_ = 'lister-item-header').find('a').text
            except:
                title = 'Not Available'
                
                
            try:
                year = box.find('span',class_ = 'lister-item-year text-muted unbold').text
            except:
                year = 'Not Available'
                
            
            try:
                genre = box.find('p',class_ = 'text-muted').find('span',class_= 'genre').text
                genre = genre.rstrip()
                genre = [genre]
            except:
                genre = 'Not Available'
            #print(genre)
           # print(type(genre))
    

            try:
                desc = box.find('div',class_ = 'lister-item-content').findAll('p')[1].text
                desc = desc.rstrip()
                #print(desc)
                desc = [desc]
            except:
                desc = 'Not Available'
            
            
            try:
                star = box.find('p',class_="").text
                star = star.rstrip()
                #print(star)
                star = [star]
            except:
                star = 'Not Available'
           # print('Title: ',title, end='\n')
            #print('Year: ',year,end='\n')
           # print('Genre: ',genre,end='\n')
           # print('Description: ',desc,end='')
           # print('starring: ',star,end="")



            try:
                certi = box.find('span',class_='certificate').text
                #print('certificate: ', certi, end='\n')
            except:
                certi = 'Yet to be certified'
                #print('Yet to be certified')


            try:
                run = box.find('span',class_='runtime').text
                #print('runtime: ',run,end='\n')
            except:
                run = 'Yet to be announced'
                #print('Yet to be announced')

                
            try:
                rate = box.find('div',class_='inline-block ratings-imdb-rating').find('strong').text
                #print('Rating: ',rate,end='\n')
            except:
                rate = 'Yet to be rated'
                #print('Yet to be rated')

                
            try:
                votes = box.find('p',class_='sort-num_votes-visible').findAll('span')[1].text
                #print('Votes: ',votes,end='\n')
            except:
                votes = 'No votes'
                #print("No votes")

            #print('\n\n')
            
            csv_writer.writerow([title,year,genre,desc,star,certi,run,rate,votes])   #Writing the data into the file
            
            # code to adjust with the html code for each page
        i += 1
        if i == 1:
            url = soup.find('div',class_ = 'desc').find('a')['href']
            print(url)
            Total += 50
            print(Total)
        else:
            try:
                url= soup.find('a',class_='lister-page-next next-page')['href']
                print(url)
                Total += 50
                print(Total)
            except:
                print("End of Scrape")
                break
        html = urlopen('https://www.imdb.com/'+url)
    csv_file.close()
    
webpage('/search/title/?title_type=movie&genres=sci-fi&explore=title_type,genres&ref_=adv_prv')
```

# Data cleaning

```python
data = pd.read_csv('D:\jupyter lab\Imdb_movie_database2.csv')   # Reading the data to the file
```

```python
data = data.dropna(subset=['Title'])
data['Title'].count()   #getting the no.of rows
```

```python
data.reset_index(drop=True,inplace=True)    # ordering the index
```

```python
data.head(50)    # viewing the data
```

```python
data['Year'].unique()    #viewing unique dates
```

```python
data = data.dropna(subset=['Year'])   # dropping NA values
```

```python
data['Year'] = data['Year'].str.extract(r"(\d\d\d\d)")
data.head()
```

```python
data.iloc[4,3]
```

using regular expressions to clean the rest of the data by pulling the required fields.

```python
data['Genre'] = data['Genre'].str.extract(r"([^\[\'\\n][a-zA-z, -]*)")
data.head()
```

```python
data['Description'] = data['Description'].str.extract(r'([^\[\"\'\\n].*[^$\]\'])')
data['Description'] = data['Description'].str.lstrip()
#data.iloc[1000,3]
```

```python
data.head()
```

```python
data.iloc[4,4]
```

```python
data['Starring'] = data['Starring'].str.extract(r'([^\[\'\"\\n].*[^$\'\]])')
data['Starring'] = data['Starring'].str.lstrip()
data[['Director','Cast']] = data['Starring'].str.split('|',expand=True)
data.drop(columns=['Starring'])
cols = list(data.columns.values)
data = data[cols[0:4]+cols[9:11]+cols[5:9]]
data.iloc[4,4]
data.head()
```

```python
data['Director'] = data['Director'].str.extract(r'([^Director:|Directors:|\\n].*[^$\\n])')
data.head()
```

```python
data.iloc[4,5]
```

```python
data['Cast'] = data['Cast'].str.extract(r'([^\\n][^ Stars:][^\\n].*)')
data.head()
```

```python
data['Cast'] = data['Cast'].str.replace(r'\\n','')
data.head()
```

```python
data['Director'] = data['Director'].str.replace(r'\\n','')
data.head()
```
